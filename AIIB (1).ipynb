```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup

pdfs = []
#Define and Read source URL
my_url = 'https://www.aiib.org/en/projects/approved/index.html'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()

#parse HTML
page_soup = soup(page_html, "html.parser")

#find all links in page
links = [a['href'] for a in page_soup.find_all('a', href=True)]

#filter links which are to project description pages
for link in links:
    if "projects/approved" in link:
        zurr = "https://www.aiib.org"
        muhurl = (zurr+link)
#run beautifulsoup on each relevant links
        bluClient = uReq(muhurl)
        blupage_html = bluClient.read()
        bluClient.close()
        blupage_soup = soup(blupage_html, 'lxml')
#extract specific formatting in which summary pdf's are linked
        links2 = blupage_soup.find_all('div', {'class': ['detail-download']})
#parse tags to isolate and extract pdf links
        z = str(links2)
        x = z.split("<")
        for a in x:
            if "a href" in a:
                b = a.split('"/')
                c = b[1].replace('" target="_blank">', '')
                d = c.replace("en/projects/approved","https://www.aiib.org/en/projects/approved")
                pdfs.append(str(d)[:-1])
                
pdfs
```

```python
import requests
import re
```

```python
for i in pdfs:
    myfile = requests.get(i)
    index = [m.start() for m in re.finditer('/', i)][-1]
    address = '/Users/tiangewan/Documents/Project/'+i[index+1:]
    open(address,'wb').write(myfile.content)
```
