```python
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd
import dash_auth
import json
import numpy as np


df=pd.read_csv('final_table_worldbank.csv', index_col=0)
df.sort_values(by=['Project Title'],ascending=False,inplace=True)
app=dash.Dash()

USERNAME_PASSWORD_PAIRS=[['tiange','wan']]
auth=dash_auth.BasicAuth(app,USERNAME_PASSWORD_PAIRS)
graphoptions=["equirectangular" ,"mercator" , "orthographic" , "natural earth" ,
              "kavrayskiy7" ,"miller" , "robinson" , "eckert4" , "azimuthal equal area" ,
              "azimuthal equidistant" , "conic equal area" ,"conic conformal" , "conic equidistant" ,"gnomonic" ,
              "stereographic" , "mollweide" , "hammer" , "transverse mercator" ,"albers usa" ,"winkel tripel" , "aitoff" , "sinusoidal" ]
project_options=[]
for project in df['Project Title']:
    project_options.append({'label':project,'value':project})

app.layout = html.Div([

    html.Div([
        dcc.Dropdown(id='project_pick',options= project_options, value= 'vaccines and cold chain equipment'),
    ],style={'width':"48%",'display':'inline-block'}),
    html.Div([
        dcc.Dropdown(id='choose_plot',options=[{'label':i,'value':i} for i in graphoptions],value=graphoptions[3])
    ],style={'width':"48%",'display':'inline-block'}),
    html.Div([dcc.Graph(id='graph')],style={'width':'100%', 'float':'left'}),
    html.Div([dcc.Textarea(id = 'text', value='This is a TextArea component',draggable = True)],style={'width':'100%'})
],style={'padding':10})

@app.callback(Output('graph', 'figure'),
              [Input('project_pick', 'value'),
               Input('choose_plot','value'),
               ]
              )
def update_figure (selected_project,selected_graph):
    filter_df= df[df['Project Title']==selected_project]

    data = dict(
        type='choropleth',
        locations=filter_df['Country'],
        locationmode='country names',
        z=filter_df['Commitment Amount'],
        text=filter_df['Country'],

        marker=go.choropleth.Marker(
            line=go.choropleth.marker.Line(
                color='rgb(180,180,180)',
                width=0.5
            )),
    
     reversescale=False,
        colorbar={"thickness": 10, "len": 0.7, "x": 0.9, "y": 0.7,
                  'title': {"side": "bottom","size":1}}
    )
    layout = dict(
        width= 900,
        height=600,
        title = 'Details for '+ selected_project,
        geo=dict(

            projection={'type': selected_graph
                        }
        )
    )
    return {'data':[data],"layout":layout}

@app.callback(Output('text', 'value'),
              [Input('project_pick', 'value')])
def update_text(value):
    filter_df= df[df['Project Title']==value]
    content = 'Project Name: '+str(filter_df['Project Title'].values[0])+ '\n'+ 'Country: '+str(filter_df['Country'].values[0])+ '\n' + 'Approved Amount: '+str(filter_df['Commitment Amount'].values[0])+ ' Million Dollars'+'\n'+'Approval Date: '+ str(filter_df['Approval Date'].values[0])
    return content
    
if __name__ == '__main__':
    app.run_server()

```
