```python
df3 = pd.read_csv('Carribean.csv')
df3
```

```python
def converter(money):
    if '('in money:
        return int(money[-16:-8].replace('$','').replace(',',''))
    if'$'in money:
        return int(money.replace('$','').replace(',',''))
    else:
        if "." in money:
            return int(money[:-3].replace('£','').replace(',',''))*1.22
        return int(money.replace('£','').replace(',',''))*1.22
```

```python
df3['approved_total'] = df3['approved_total'].apply(converter).values
```

```python
updated_carribean = df3[['date','country','approved_total']]                      
```
