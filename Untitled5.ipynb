```python
from urllib.request import urlopen
from urllib.request import Request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
main = 'https://www.bcie.org/operaciones-y-adquisiciones/operaciones-aprobadas/?no_cache=1'
```

```python
req = Request(main,headers={'User-Agent':'Mozilla/5.0'})
client = urlopen(req)
page = client.read()
client.close()
soup = BeautifulSoup(page,"html.parser")
```

```python
information = soup.find_all('td')
```

```python
numbers = [information[i].text[-4:] for i in range(0,len(information),8)]
numbers
```

```python
countries = [information[i].text[6:] for i in range(1,len(information),8)]
countries
```

```python
areas = [information[i].text[22:] for i in range(2,len(information),8)]
areas
```

```python
projects = [information[i].text[24:].replace('\r','').replace('\n','') for i in range(3,len(information),8)]
projects
```

```python
components = [information[i].text[12:] for i in range(4,len(information),8)]
components
```

```python
categories = [information[i].text.replace('\t','').replace('\n','')[29:] for i in range(5,len(information),8)]
categories
```

```python
investments = ['US '+ information[i].text.replace('\n','').replace('\xa0','')[35:] for i in range(6,len(information),8)]
investments
```

```python
dates = [information[i].text[21:-1] for i in range(7,len(information),8)]
dates
```

```python
df = pd.DataFrame({"Línea de Crédito":numbers,"País":countries,"Área de Focalización":areas,"Nombre de la Operación":projects,"Componente":components,"Categoría Ambiental y Social":categories,"Monto aprobado por el BCIE (USD)":investments,"Fecha de Aprobación":dates})
df
```
