```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
features = ['a', 'b', 'c', 'd']
features
```

```python
sc = {}
for i in features:
    mean,std = df[i].mean(),df[i].std()
    sc[i]=[mean,std]
    df.loc[:,i]=(df[i]-mean)/std
    
```

```python
df
```

```python
target = df.values[:,4]
target
```

```python
features = df.values[:,:4]
features
```

```python
feature_train,feature_test,target_train,target_test = train_test_split(features,target,test_size=0.2)
```

```python
clf = GaussianNB()
```

```python
clf.fit(feature_train,target_train)
target_pred = clf.predict(feature_test)
```

```python
accuracy_score(target_test,target_pred,normalize = True)
```
